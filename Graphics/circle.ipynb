```python
import matplotlib.pyplot as plt
import numpy as np

# Define the center of the circle and its radius
cx, cy, r = 0, 0, 1

# Define the number of points
n = 100

# Generate the points
theta = np.linspace(0, 2*np.pi, n)
x = cx + r * np.cos(theta)
y = cy + r * np.sin(theta)

# Plot the points
plt.scatter(x, y)
plt.gca().set_aspect('equal', adjustable='box')  # To ensure the circle isn't distorted
plt.show()
```
