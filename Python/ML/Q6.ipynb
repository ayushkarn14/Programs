```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

data = pd.read_csv('Advertising.csv')

correlation = data[['TV', 'Radio', 'Newspaper', 'Sales']].corr()


print(correlation['Sales'])
```

```python
x = data['TV'].values
y = data['Sales'].values


A = np.vstack([x, np.ones(len(x))]).T
m, c = np.linalg.lstsq(A, y, rcond=None)[0]

y_pred = m * x + c

plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.title('Line of Regression: TV vs Sales')
plt.show()

# Calculate the error
error = np.sum((y - y_pred) ** 2)
print('Error:', error)
```
