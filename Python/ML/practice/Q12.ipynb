Consider the dataset iris. Apply the PCA method to select the best 2 features. Using these
features plot the scatter graph. Apply k-means clustering algorithm to cluster the transformed
dataset into 3 clusters.


```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import pandas as pd
```

```python
iris=load_iris()
x=iris.data
y=iris.target
```

```python
pca=PCA(n_components=2)
x=pca.fit_transform(x)
```

```python
model=KMeans(n_clusters=3)
a=model.fit_transform(x)
a.shape
```

```python
import matplotlib.pyplot as plt
plt.scatter(x[:,0],x[:,1])
```

```python
model.labels_
```

```python
df=pd.DataFrame(x,model.labels_)
```

```python
df['Class']=model.labels_
```

```python
df.columns=['F1','F2','Class']
```

```python
plt.scatter(df['F1'],df['F2'],c=df['Class'])
```
