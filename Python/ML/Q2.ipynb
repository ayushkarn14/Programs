Classify the iris dataset using a Bayes classifier. Divide the dataset into training and testing in the ratio 80:20. Use the functions from the sklearn package. Assume the data follows a
gaussian distribution. Display the training and testing accuracy, confusion matrix.

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
iris=load_iris()
x=iris.data
y=iris.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.2,random_state=0)
```

```python
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
classifier=GaussianNB()
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
y_pred
```

```python
y_test
```

```python
cm=confusion_matrix(y_pred=y_pred,y_true=y_test)
print(cm)
```

```python
sns.heatmap(cm,annot=True,fmt='.2f',xticklabels=['setosa', 'versicolor', 'virginica'],yticklabels=['setosa', 'versicolor', 'virginica'])
```
