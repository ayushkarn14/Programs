Classify the iris dataset using a decision tree classifier. Divide the dataset into training and
testing in the ratio 80:20. Use the functions from the sklearn package. Display the final decision
tree.

```python
from sklearn.datasets import load_iris
iris=load_iris()
x=iris.data
y=iris.target
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
model=DecisionTreeClassifier() #seeing the depth of tree without setting the max depth first
model.fit(x_train,y_train)
print("Tree depth : ",model.get_depth()) #as depth is 5 for 4 features, we don't prune the tree
a=tree.plot_tree(model,filled=True)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```
