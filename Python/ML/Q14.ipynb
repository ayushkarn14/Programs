Consider the dataset iris. Apply hierarchical clustering algorithm to cluster the dataset into 3
clusters.

```python
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering

iris = load_iris()
X = iris.data

iris
```

```python
n_clusters = 3
aggmodel = AgglomerativeClustering(distance_threshold=None,n_clusters=n_clusters,linkage = "complete")
labels = aggmodel.fit_predict(X)

aggmodel.n_clusters_
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 1],c=labels)
plt.xlabel('Sepal Width')
plt.ylabel('Petal Length')
plt.title('Cluster Plot')
plt.show()
```

```python
plt.scatter(X[:, 0], X[:, 2])
plt.xlabel('Sepal Width')
plt.ylabel('Petal Width')
plt.title('Cluster Plot')
plt.show()
```

```python
plt.scatter(X[:, 1], X[:, 2], c=labels)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Cluster Plot')
plt.show()
```
