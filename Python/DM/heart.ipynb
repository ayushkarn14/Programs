```python
import pandas as pd
import seaborn as sns
```

```python
heart=pd.read_csv("heart.csv")
heart.head()
```

```python
women=heart.loc[heart['Sex']=='F']['HeartDisease']
women
```

```python
rate_women=(women.sum()/len(women)).round(2)*100
print("Percentage of Women with probability of Heart Disease = ", rate_women, "%")
```

```python
men=heart.loc[heart['Sex']=='M']['HeartDisease']
rate_men=(men.sum()/len(men)).round(2)*100
print("Percentage of Women with probability of Heart Disease = ", rate_men, "%")
```

```python
heart.agg({
    "Age":["min","max","median","skew","std"],
    "RestingBP":["min","max","median","skew","std"],
    "Cholesterol":["min","max","median","skew","std"],
    "Oldpeak":["min","max","median","skew","std"],
    "MaxHR":["min","max","median","skew","std"]
})
```

```python
sns.kdeplot(data=heart,x="Cholesterol",hue="ChestPainType",fill=True)
```

```python
sns.displot(data=heart,x="Cholesterol",hue="ChestPainType",col="Sex",kind="kde")
```
